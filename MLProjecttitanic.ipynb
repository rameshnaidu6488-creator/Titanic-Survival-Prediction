import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data collecting & processing


```python
td=pd.read_csv("/content/train.csv")
```

```python
td.head()
```

```python
td.shape
```

```python
td.info()
```

```python
td.isnull().sum()
```

```python
td=td.drop(columns="Cabin",axis=1)
```

```python
td["Age"].fillna(td["Age"].mean(),inplace=True)
```

```python
print(td["Embarked"].mode())
```

```python
td['Embarked'].fillna(td['Embarked'].mode()[0],inplace=True)
```

```python
td['Embarked'].value_counts()
```

encoding the categorical columns

```python
td.replace({"Sex":{"male":0,"female":1},"Embarked":{'S':0,'C':1,'Q':2}},inplace=True)
```

```python
td.head()
```

separating features

```python
x=td.drop(columns=["PassengerId",'Name','Ticket','Survived'],axis=1)
y=td['Survived']
```

```python
print(x)

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_trainp=model.predict(x_train)
```

```python
print(x_trainp)
```

```python
acc_score=accuracy_score(y_train,x_trainp)
```

```python
acc_score
```

```python
y_testp=model.predict(x_test)
acc_scoret=accuracy_score(y_test,y_testp)
```
